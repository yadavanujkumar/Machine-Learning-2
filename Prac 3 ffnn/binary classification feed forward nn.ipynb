```python
import pandas as pd
```

```python
df = pd.read_csv("pima.csv", header=None)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classiffnn = Sequential()
```

```python
classiffnn.add(Dense(10, input_dim=8, activation='relu'))
classiffnn.add(Dense(8, activation='relu'))
classiffnn.add(Dense(1, activation='sigmoid'))
```

```python
classiffnn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
classiffnn.fit(xtrain, ytrain, epochs=500, batch_size=20)
```

```python
predictions = classiffnn.predict(xtest)
```

```python
predictions
```

```python
class_label=[]
for prediction in predictions:
  if prediction[0]>0.5:
    class_label.append(1)
  else:
    class_label.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(ytest, class_label)
```
