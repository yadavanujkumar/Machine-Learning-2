```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ml 2/movie.csv')
df
```

```python
df.columns
```

```python
df.shape
```

```python
features = ['genres','keywords','original_language','title','cast','director']
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
def combined_features(row):
  return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']+" "+row['original_language']+" "+row['title']

df["combined_features"]=df.apply(combined_features,axis=1)
df["combined_features"]
```

```python
tfid = TfidfVectorizer()
tfidv = tfid.fit_transform(df["combined_features"])
tfidv.toarray()
```

```python
tfidv.shape
```

```python
cosine_sim = cosine_similarity(tfidv)
cosine_sim.shape
```

```python
movie = (input("Enter movie name: "))

def get_index(mn):
  return df[df.title == mn]["index"].index[0]

mi = get_index(movie)
mi
```

```python
similarity_scores = list(enumerate(cosine_sim[mi]))
sorted_similar_movies = sorted(similarity_scores,key=lambda x:x[1],reverse=True)
i=0
print('Top 10 Recommended Movies are : \n')
for element in sorted_similar_movies:
    print(df[df.index == element[0]]['title'].values[0])
    i=i+1
    if i>10:
        break
```
