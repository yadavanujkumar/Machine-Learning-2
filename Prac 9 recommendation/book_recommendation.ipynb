```python
import pandas as pd
books = pd.read_csv("/content/BX-Books - BX-Books (1).csv" , encoding = "latin-1",delimiter=",")
books
```

```python
ratings = pd.read_csv("/content/BX-Book-Ratings.csv" , encoding="latin-1",delimiter=";")
ratings
```

```python
rating_books = pd.merge(ratings,books,on="ISBN")
rating_books
```

```python
rating_books.shape
```

```python
rating_books_sample = rating_books.sample(frac=0.1, random_state=1)
rating_books_sample.sample(100)
```

```python
rating_books_pivot = rating_books_sample.pivot_table(index=["Book-Title"],columns=["User-ID"],values="Book-Rating").fillna(0)
rating_books_pivot
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric="cosine",algorithm="brute",n_neighbors=7,n_jobs=-1)
model.fit(rating_books_pivot)
```

```python
indices= model.kneighbors(rating_books_pivot.loc[["Tears of Rage"]],10,return_distance=False)
```

```python
import numpy as np

for index,values in enumerate(rating_books_pivot.index[np.array(indices).flatten()]):
  print((index+1),".",values)
```
