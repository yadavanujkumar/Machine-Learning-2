```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset=load_iris()
```

```python
x=dataset.data
y=dataset.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = AdaBoostClassifier(n_estimators=50)
model.get_params()
```

```python
model.fit(x_train,y_train)
```

```python
predictab=model.predict(x_test)
```

```python
predictab
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of Adaboost = ", accuracy_score(y_test,predictab))
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ml 2/train.csv")
test_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ml 2/test.csv")
```

```python
print(test_data.head())
```

```python
# The columns in drop_columns were already removed in previous steps.
drop_columns = ['Name', 'Ticket', 'Cabin', 'Embarked','Parch', 'Age']
train_data_clean = train_data.drop(drop_columns, axis=1 , inplace=True)
test_data_clean = test_data.drop(drop_columns, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_data['Sex'] = le.fit_transform(train_data['Sex'])
test_data['Sex'] = le.fit_transform(test_data['Sex'])
```

```python
ytrain = train_data['Survived']
train_data.drop('Survived', axis=1, inplace=True)
xtrain = train_data
```

```python
from sklearn.model_selection import train_test_split

xtrain,xval,ytrain,yval=train_test_split(xtrain,ytrain,test_size=0.3,random_state=12)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
# Remove rows with missing values

lr_list = [0.05,0.075,0.1,0.25,0.5,0.75,1]

for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=20,
                                        learning_rate=learning_rate,
                                        max_features=2, max_depth=2, random_state=0)
    # Fit the model on the cleaned data
    gb_clf.fit(xtrain, ytrain)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(xtrain, ytrain)))
    print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(xval, yval))) # Also evaluate on validation set
```

```python
gb_clf2 = GradientBoostingClassifier(n_estimators=20,
                                        learning_rate=0.5,
                                        max_features=2, max_depth=2, random_state=0)
# Remove rows with missing values before fitting

gb_clf2.fit(xtrain, ytrain)
```

```python
predections=gb_clf2.predict(xval)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy score of GradientBoost = \n",accuracy_score(yval,predections))
```

```python
print("Confusion matrix of GradientBoost = \n",confusion_matrix(yval,predections))
```

```python
print("Classification report of GradientBoost = \n",classification_report(yval,predections))
```
