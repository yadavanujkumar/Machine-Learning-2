comparing kmeans and gasussianmixture algo


```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ml 2/Clustering_gmm - Clustering_gmm.csv')
```

```python
dataset.head(10)
```

```python
plt.figure()
plt.scatter(dataset['Weight'], dataset['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(dataset)
```

```python
predictions = kmeans.predict(dataset)
```

```python
predictions
```

```python
dataframe1=pd.DataFrame(dataset)
dataframe1
```

```python
dataframe1['predictions']=predictions
dataframe1
```

```python
color=['red','yellow','green','blue']
for i in range(0,4):
  data=dataframe1[dataframe1['predictions']==i]
  plt.scatter(data['Weight'],data['Height'],c=color[i])
plt.show()
```

```python
dataset_gmm=dataset[['Weight','Height']]
```

```python
gmm=GaussianMixture(n_components=4)
gmm.fit(dataset_gmm)
```

```python
pred=gmm.predict(dataset_gmm)
```

```python
dataframe2=pd.DataFrame(dataset_gmm)
dataframe2
```

```python
dataframe2['predictions']=pred
dataframe2
```

```python
color=['red','yellow','green','blue']
for i in range(0,4):
  data=dataframe2[dataframe2['predictions']==i]
  plt.scatter(data['Weight'],data['Height'],c=color[i])
plt.show()
```
